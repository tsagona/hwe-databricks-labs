# Week 1 Lab: Exploring Data with SQL in Databricks

In this lab you'll write your first queries against a real dataset using the Databricks SQL editor.

We'll use the **NYC Taxi** sample dataset that Databricks ships with every workspace. It lives in Unity Catalog under `samples.nyctaxi.trips` — no setup required.

## Step 1: Browse the Data

Let's start simple: pull back all columns from the table to see what we're working with.

The `LIMIT` clause caps the result at 10 rows so we don't wait for a full table scan.

```sql
SELECT *
FROM samples.nyctaxi.trips
LIMIT 10
```

---
## Step 2: Aggregate by Pickup Zip Code

Now let's ask a more interesting question: which pickup areas generate the most rides and the highest fares?

This query groups trips by pickup zip code, counts the rides, and computes average and total fare — then sorts by total revenue descending so the busiest zones float to the top.

```sql
SELECT
  pickup_zip,
  COUNT(*)                          AS total_trips,
  ROUND(AVG(fare_amount), 2)        AS avg_fare,
  ROUND(SUM(fare_amount), 2)        AS total_fare,
  ROUND(AVG(trip_distance), 2)      AS avg_distance_miles
FROM samples.nyctaxi.trips
GROUP BY pickup_zip
ORDER BY total_fare DESC
LIMIT 20
```
