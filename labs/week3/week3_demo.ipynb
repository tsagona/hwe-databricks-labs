# Week 3 Demo: Writing Your First Tests

This notebook demonstrates two testing patterns:
- **SELECT COUNT**: run a query and assert the returned count
- **INSERT + verify**: run an INSERT, then SELECT to verify the result

Each cell is tagged with `-- @test:tag_name` so the test file can find and run it.
